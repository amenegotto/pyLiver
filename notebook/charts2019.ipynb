```python
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_color_codes("pastel")
```

```python
obj_df = pd.read_csv('/home/amenegotto/Downloads/Artigos Coletados - Objetivos.csv')
obj = sns.barplot(data= obj_df, color='black',
    y= 'Objetivo',
    x= 'Qtd')
obj.set_xlabel('')
obj.set_ylabel('')
obj_fig = obj.get_figure()
obj_fig.savefig('/tmp/chart_objectives.eps', format='eps', bbox_inches='tight')
```

```python
import numpy as np

org_df = pd.read_csv('/home/amenegotto/Downloads/Artigos Coletados - Orgaos.csv')

color = plt.cm.gray(np.linspace(0.3, 0.9, 13))
plt.pie(org_df['Qtd'], labels=org_df['Orgao'], colors=color)

#plt.set_xlabel('')
#plt.set_ylabel('')
plt.savefig('/tmp/chart_orgaos.eps', format='eps', bbox_inches='tight')
```

```python
inputs_df = pd.read_csv('/home/amenegotto/Downloads/Artigos Coletados - Inputs.csv')
sns.set_palette("hls")
ax = sns.heatmap(inputs_df.corr(), annot=True, cmap="Greys")
fig = ax.get_figure()
fig.savefig('/tmp/correlation_inputs.eps', format='eps', bbox_inches='tight')
```

```python
inputs_df = pd.read_csv('/home/amenegotto/Downloads/Artigos Coletados - Inputs.csv')
color = plt.cm.gist_yarg(np.linspace(0.2, 0.8, 7))
inputs_df.plot(kind='bar', stacked=True, color=color)

#plt.set_xlabel('')
plt.xticks([])
#plt.set_ylabel('')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.savefig('/tmp/chart_multi_inputs.eps', format='eps', bbox_inches='tight')
```
