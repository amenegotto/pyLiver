```python
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_color_codes("pastel")
```

```python
obj_df = pd.read_csv('/home/amenegotto/Downloads/Resultados - BaseDiag.csv')
obj = sns.barplot(data= obj_df, color='black',
    y= 'Tecnica',
    x= 'Quantidade')
obj.set_xlabel('')
obj.set_ylabel('')
obj_fig = obj.get_figure()
obj_fig.savefig('/tmp/chart_diagbase.eps', format='eps', bbox_inches='tight')
```
